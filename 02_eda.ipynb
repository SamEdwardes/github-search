```python
import pandas as pd
import os
import re

from sklearn.feature_extraction.text import CountVectorizer
```

```python
pd.set_option('display.max_rows', 500)
```

```python
def get_extension(x):
    path, ext = os.path.splitext(x)
    return ext

get_extension("archive/2018-19/lectures/S4.pdf")
```

```python
df = pd.read_csv("data/2020-02-27_student-repos.csv")
df["ext"] = df["path"].apply(get_extension)
df
```

```python
len(df[df["content"] == "not read"])/len(df)
```

```python
df["ext"].value_counts()
```

```python
df = df[df["content"] != "not read"]
df
```

```python
df["ext"].value_counts()
```

```python
def text_preprocess(x):
    regex = re.compile('[^a-zA-Z ]')
    x = regex.sub('', x)
    return x
```

```python
df["content_clean"] = df.loc[:,"content"].apply(text_preprocess)
```

```python
df["content_clean"]
```

```python
count_vec = CountVectorizer(stop_words="english")
X = count_vec.fit_transform(df["content_clean"])
```

```python
pd.DataFrame(data=X.toarray(), columns=count_vec.get_feature_names())
```

```python
pd.DataFrame(data={
    "word": count_vec.get_feature_names(),
    "count":pd.DataFrame(X.sum(axis=0)).T.to_numpy().ravel()
}).sort_values(by="count", ascending=False).head(100)
```
